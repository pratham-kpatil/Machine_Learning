## Q1. Apriori Algorithm Cosmetics Dataset

```python
#importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
#importing data
data = pd.read_csv(r'C:\Users\patil\OneDrive\Desktop\JYPN\Datasets\Cosmetics.csv')
data.head()
```

```python
#replacing Yes with 1 and No with 0
data.replace({'Yes':1, 'No':0}, inplace=True)
data.head()
```

### Generating Rules with support>25% and confidence>70%

```python
#Generating frequent itemset
freq_items = apriori(data,min_support=0.001 , use_colnames=True)
freq_items
```

```python
#Genrating the rules from frequent datasets
rules = association_rules(freq_items, metric='confidence', min_threshold=0.001)
rules = rules[rules.lift>=1]
rules.head()
```

```python
print(rules.shape)
rules.describe()
```

### Sorting rules having lift geater than 1

```python
rules = rules[rules.lift>1]
print(rules.shape)
```
