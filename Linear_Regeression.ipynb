```python
data = np.loadtxt(os.path.join('Profit.txt'), delimiter=',')
X, y = data[:, 0], data[:, 1]
m = y.size  
print(m)
```

```python
def plotData(x, y):
    fig = pyplot.figure() 
    pyplot.plot(X, y, 'ro', ms=10, mec='k')
    pyplot.ylabel('Profit in $10,000')
    pyplot.xlabel('Population of City in 10,000s')
```

```python
plotData(X, y)
```

```python
X = np.stack([np.ones(m), X], axis=1)
```

```python
def computeCost(X, y, theta):
    m = y.size  
    J = 0
    print(theta)
    J = (1/(2*m))*(np.sum(np.square(np.dot(X,theta)-y)))
    
    return J
```

```python
A = 0.69
B = -0.42
J = computeCost(X, y, theta=np.array([A, B]))
print('With theta = [-1,2] \nCost computed = %.3f'%J)
```

```python
def gradientDescent(X, y, theta, alpha, num_iters):
    
    m = y.shape[0] 
    theta = theta.copy()
    J_history = []
    
    for i in range(num_iters):
        temp0 = theta[0] -(alpha/m)*(np.sum((np.dot(X,theta))-y))
        temp1  = theta[1] - (alpha/m)*(np.sum((np.dot(X,theta)-y)*X[:,1]))
        theta[0] = temp0
        theta[1] = temp1
        J_history.append(computeCost(X, y, theta))
    
    return theta, J_history
```

```python
A = 2
B = -1
theta = np.array([A,B], dtype=np.float32) 

iterations = 1500
alpha = 0.001

theta, J_history = gradientDescent(X ,y, theta, alpha, iterations)
print('Theta found by gradient descent: {:.4f}, {:.4f}'.format(*theta))
```

```python
plotData(X[:, 1], y)
pyplot.plot(X[:, 1], np.dot(X, theta), '-')
pyplot.legend(['Training data', 'Linear regression']);
```
